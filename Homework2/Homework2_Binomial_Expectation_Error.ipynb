```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns 
import math
```

```python
def InrangeProbability(bino,n,p,size,error):
    count = 0
    exp = n*p
    a = math.floor(exp*(1-error))
    b = math.ceil(exp*(1+error))
    for x in bino:
        if (x >= a) and (x <= b):
            count+=1
    #print(count)
    return count/size
```

```python
def DoSimulation(n,p,e):
    size = 100000
    b = random.binomial(n,p,size)
    print(b)
    sns.distplot(b)
    plt.show()
    answer = InrangeProbability(b,n,p,size,e)
    print(answer)
    return answer
```

```python
b00 = round(DoSimulation(100,0.25,0.01),4)
b00
```

```python
b10 = round(DoSimulation(3000,0.25,0.01),4)
b10
```

```python
b20 = round(DoSimulation(50000,0.25,0.01),4)
b20
```
