```python
import numpy as np
import pylab as pl
from scipy.special import gamma
```

```python
def binomial_distribution_plot(n,p):
    x = np.linspace(1,n+1,100)
    Probability_part = (p**(x-1))*((1-p)**(n-(x-1)))
    Combanation_part = gamma(n+1) / (gamma(x)*gamma(n-x+2))
    
    pl.plot(x-1,Probability_part, label='probability_n='+str(n)+'_p='+str(p))
    pl.plot(x-1,Combanation_part, label='combination_n='+str(n)+'_p='+str(p))
    pl.plot(x-1,Probability_part*Combanation_part, label='binomial_n='+str(n)+'_p='+str(p))
    pl.legend(loc='best')
    pl.xlabel('x')
    pl.ylabel('y')
    pl.rcParams["figure.figsize"] = (8, 6)
```

```python
def log10_binomial_distribution_plot(n,p):
    x = np.linspace(1,n+1,100)
    Probability_part = (p**(x-1))*((1-p)**(n-(x-1)))
    Combanation_part = gamma(n+1) / (gamma(x)*gamma(n-x+2))
    
    k = (x-1)/n
    pl.plot(k,np.log10(Probability_part), label='probability_n='+str(n)+'_p='+str(p))
    pl.plot(k,np.log10(Combanation_part), label='combination_n='+str(n)+'_p='+str(p))
    pl.plot(k,(np.log10(Probability_part)+np.log10(Combanation_part)), label='binomial_n='+str(n)+'_p='+str(p))
    pl.legend(loc='best')
    pl.xlabel('x')
    pl.ylabel('(log10)y')
    pl.rcParams["figure.figsize"] = (16, 12)
```

```python
binomial_distribution_plot(10,0.5)
```

```python
binomial_distribution_plot(30,0.7)
```

```python
log10_binomial_distribution_plot(10,0.5)
log10_binomial_distribution_plot(30,0.7)

pl.show()
```
